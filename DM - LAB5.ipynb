```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
#Let's get the data ready 

Freq_itemset_sizes = list(range(2, 16)) #get all frequent itemsets from size 2 to 15.
minsup = [5, 10, 20] #the mimimum support threshold tested with the apriori algorithem

#Returned values from Apriori Algorithm ran with prompt
data = {
    5: [1329, 10618, 48226, 144928, 315873, 527176, 692740, 723735, 600196, 391578, 197889, 75624, 21041, 4000],
    10: [763, 4593, 16150, 38800, 69835, 98846, 111786, 100660, 71342, 39171, 16292, 4956, 1039, 134],
    20: [376, 1472, 3559, 6267, 8802, 10151, 9488, 7010, 4004, 1729, 546, 119, 16, 1]
}
```

```python
#Let's plot


for thresh in minsup:
    plt.figure() 
    plt.plot(Freq_itemset_sizes, data[thresh], marker='o')
    plt.xlabel('Frequent Itemset Size')
    plt.ylabel('Number of Frequent Itemsets')
    plt.title(f'Number of Frequent Itemsets (MinSup {thresh})')
    plt.grid(True) #for better visualization
    plt.savefig(f'frequen itemsets, minsup={thresh}.PNG')


plt.show()
```

```python
#Let's plot them all on one graph 
for thresh in minsup:
    plt.plot(range(2, 16), data[thresh], label=f'MinSup{thresh}')


plt.xlabel('Frequent Itemset Size')
plt.ylabel('Number of Frequent Itemsets')
plt.title('Number of Frequent Itemsets by Support Threshold')
plt.legend()
plt.savefig('Freq Itemsets all-in-one')

plt.show()
```
